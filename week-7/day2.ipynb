# Streamlit


### Main Concepts


#### Development Flow

Streamlit provides a fast interactive loop. Everytime there is a change in the source code, the streamlit detects and asks you if you want to rerun the app to refelct the chaanges.

- Choose Always rerun to automatically update the app.


#### Data Flow

- streamlit reruns the entire python script when:
1. you modify the apps source code.
2. when user interacts with the widgets in the app.

**Note:** Streamlit uses `@st.cache_data` decorator to optimize the reruns.

#### Write

- versatile command in streamlit. 
- multiple agguments can be passed. 
- the behavior depends on the input type.
- retruns none, so its slot can't be reused.

#### Magic

- Magic commands are a feature in streamlit to write anything without explicit command.
- any variable or literal in streamlit file will ba automatically writen to your app using `st.write`.
- to disable maic set `magicEnabled = false` in `~/.streamlit/config.toml`.

#### Widgets

- elemesnt of GUI.

- `st.slider`
- `st.button`
- `st.text_input`

#### Layout

- used to organize the widgets.
- `st.sidebar`


#### Themes

- streamlit supports light and dark themes
- settings --> theme -->select theme
- streamlit also provides an option to select a custom theme

#### Pages

- streamlit supports frictionless way to create multiple pages.
- create a folder pages in the directory main page ans add python files.

**File names**

- A number — if the file is prefixed with a number.
- A separator — could be _, -, space, or any combination thereof.
- A label — which is everything up to, but not including, .py.
- The extension — which is always .py.


### Advanced Topics

#### Session State

session state provides a dictionary like interface to store values between reruns.

syntax:

`st.session_state.key` or `st.session_state['key]`


#### Caching

- streamlit caches the result of expensive operations
- to cache a result in streamlit, you need to decorate it with
    - `@st.cache_data` : to cache data
    - `@st.cache_resource`:  to cache global resources like db connections

#### Deployment


- streamlit community cloud
- custom deployment



**References**

- [https://docs.streamlit.io/](https://docs.streamlit.io)

- [cheat sheet](https://docs.streamlit.io/library/cheatsheet)

# Vertex AI

It is a machine learning (ML) platform in GCP that lets you train and deploy ML models, AI applications, and customize the LLMs for AI powered applications.

## Model Garden 

an enviorment to search, discover and interact with a wide variety of models(Open Source, Tast specific and foundation). The model can directly accessed using an API.

## Generative AI Studio

Easy to use, low code tooling for tuning & deploying foundation models. features like Prompt design, Prompt tuning and Fine tuning.

## ML Platform

Platform for custom ML model development & deployment with MLOps tools.


ML workflow:

1. Ingest 
2. Analyze
3. Transform
4. Train
5. Model
6. Evaluate
7. Deploy
8. Predict

- 1 to 3: data preperation. done using managed datasets in vertexAI. lable and annotate the data.
- train- AutoML and custom. img, vid, text and tabular - automl is suitable. custom model- tensorflow and pytorch.
- Evaluate: explainable AI, evaluate the model and dive deeper into the model and understand it better.
- deploy: deploy it to an endpoint(scalable and h/w). 
-  undeployed for batch predictions. deployed model prediction - CLI, UI, SDK and APIs


## Datasets

1. Data is added to a dataset(a central location to store, anaotate and label data, and access the data)--> train the model.

- Images
   - Classification: model predits one or many labels. Identifing the types if dogs.
   - Object Detection: Location of objects using bounding objects.
   - Segmentation: Model can assign a lable to pixel level in the image.
- Training images should be versatile and close to the usecase. suggested 1k per lable.

- Tablular
  - Regression: predicting the numerical value
  - classification: precting the category
  - Forcasting: predicting the output of something

- Text dataset
  - Classification: assign one or more lables to a certain document
  - Entity Extraction: model identifies the custom entity types. ex: pricing
  - Sentiment Analysis: emotion: +,- and neutral.

- Video
  - Classification: lable predictions for an entire video. ex: commercials and game
  - Action recognition: recognising the action. touchdown or a goal.
  - object tracking : lables, bounding boxes and the time stamps. ex: football.
